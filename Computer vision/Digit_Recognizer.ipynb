**Algorithm comparison and selection**

---



```
from sklearn.model_selection import KFold       # k-fold cross validation
from sklearn.model_selection import cross_val_score

features = df.iloc[:, 1:785]        # create the 10 folds
label = df.iloc[:, 0]
```

```
# compare a set of classification algorithms
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
```

```
models = []       # create a list, with one item per algorithm

models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('kNN', KNeighborsClassifier()))
models.append(('DT',  DecisionTreeClassifier()))
models.append(('NB',  GaussianNB()))
```

```
scoring = 'accuracy'    # the scoring function
```

```
# evaluate classifiers
results = []
names   = []
for name, model in models:
  kfold = KFold(n_splits=10, random_state=None)
  cv_results = cross_val_score(model, features, label, cv=kfold, scoring=scoring)
  results.append(cv_results)
  names.append(name)
  print("%03s: %f (+/- %f)" % (name, cv_results.mean(), cv_results.std()))
```

**k-Nearest Neighbors**

---

**Model Training**

```
from sklearn.model_selection import train_test_split        # split the dataset into training/test
train_features,test_features,train_label,test_label=train_test_split(features,label,random_state=2,train_size=0.8)
```

```
clf=KNeighborsClassifier()        # select kNN as the winning algorithm
clf.fit(train_features,train_label.values.ravel())
```

```
predictions=clf.predict(test_features)        # predict the results
```

```
from sklearn.metrics import accuracy_score        # accuracy score
print(accuracy_score(test_label,predictions))
```

**Evaluate the model**

```
from sklearn.metrics import confusion_matrix      # evaluate the model using a confusion matrix
confusion_matrix = confusion_matrix(test_label, predictions)
print(confusion_matrix)
```

```
from sklearn.metrics import classification_report     # classification report (precision, recall, f1 score and support score)
print(classification_report(test_label, predictions))
```
